## Understanding of the Variables

### Categorical Variables

#### City Column

```python
dataset.City.value_counts()
```

```python
dataset.City.value_counts(normalize = True)
```

```python
dataset.City.value_counts().plot(kind="bar")
plt.title("Value counts of the City variable")
plt.xlabel("Name of the City")
plt.xticks(rotation=0)
plt.ylabel("Count")
plt.show()
```

#### Gender Column

```python
dataset.Gender.value_counts()
```

```python
dataset.Gender.value_counts(normalize = True)
```

```python
dataset.Gender.value_counts().plot(kind="bar")
plt.title("Value counts of the Gender variable")
plt.xlabel("Gender type")
plt.xticks(rotation=0)
plt.ylabel("Count")
plt.show()
```

#### Illness Column

```python
dataset.Illness.value_counts()
```

```python
dataset.Illness.value_counts(normalize = True)
```

#### Categorical Vs Numerical

```python
corrmat = dataset.corr()
hm = sns.heatmap(corrmat, 
                 cbar=True, 
                 annot=True, 
                 square=True, 
                 fmt='.2f', 
                 annot_kws={'size': 10}, 
                 yticklabels=dataset.columns, 
                 xticklabels=dataset.columns, 
                 cmap="Spectral_r")
plt.show()
```
